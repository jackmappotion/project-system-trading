```python
import FinanceDataReader as fdr_broker
```

```python
class FDR_LOADER:
    def __init__(self, fdr_broker) -> None:
        self.fdr_broker = fdr_broker

    def get_krx_df(self):
        krx_df = self.fdr_broker.StockListing("KRX")
        return krx_df

    def get_ohlcv_df(self, symbol, start=None, end=None):
        ohlcv_df = self.fdr_broker.DataReader(symbol=symbol, start=start, end=end)
        ohlcv_df = ohlcv_df.loc[:, ["Open", "High", "Low", "Close", "Volume"]]
        return ohlcv_df
```

```python
fdr_loader = FDR_LOADER(fdr_broker)
```

```python
krx_df = fdr_loader.get_krx_df()
```

```python
stockcode = "005930"
start = "20220101"
end = None

ohlcv_df = fdr_loader.get_ohlcv_df(symbol=stockcode, start=start, end=end)
ohlcv_df.tail()
```
